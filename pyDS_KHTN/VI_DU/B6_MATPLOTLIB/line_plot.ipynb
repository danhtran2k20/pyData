```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
year = [1950, 1951, 1952, 1953, 1954]
pop = [2.53, 2.57, 2.62, 2.67, 2.71]
```

```python
# plot year and pop using default line style and color
plt.plot(year, pop)
plt.show()
```

```python
# plot year and pop using blue circle markers
plt.plot(year, pop, 'ro')
plt.show()
```

```python
# plot pop using x as index array 0..N-1
plt.plot(pop)        
plt.show()
```

```python
# ditto, but with red plusses
plt.plot(pop, 'r+')     
plt.show()
```

```python
plt.plot(year, pop, color='green', marker='o', linestyle='dashed',
        linewidth=2, markersize=12)
plt.show()
```

```python
plt.plot(year, pop, color='green', marker='o', linestyle='dashed',
        linewidth=2, markersize=12)
plt.title('Biu  dn s',color="red")
plt.ylabel('Dn s', color="red")
plt.xlabel('Nm', color="red")
plt.show()
```

```python
# Plotting multiple sets of data
plt.plot([1, 2, 3], label='line1', marker='*', markersize=10)
plt.plot([1.5, 2.5, 3.5], label='line2', marker='o', markersize=10)
plt.title('Biu  song song',color="red")
plt.ylabel('y value', color="red")
plt.legend(title="Ghi ch")
plt.grid(True)
plt.show()
```

```python
plt.plot([1, 2, 3], 'g^', [2, 4, 6], 'g-')
plt.show()
```

```python
# Plotting multiple sets of data
plt.plot([1, 2, 3], label='line1', marker='*', markersize=10)
plt.plot([1.5, 2.5, 3.5], label='line2', marker='o', markersize=10)
plt.title('Biu  song song',color="red")
plt.ylabel('y value', color="red")
plt.legend(title="Ghi ch")

plt.show()
```
