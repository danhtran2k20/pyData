```python
np.random.seed(0)
mu = 170 #mean
sigma = 6 #stddev
sample = 100
height = np.random.normal(mu, sigma, sample)
weight = (height-100) * np.random.uniform(0.75, 1.25, 100) # mean: 0.75 std: 1.25, size=100
```

```python
gym = pd.DataFrame({'height': height, 'weight': weight})
gym.head()
```

```python
gym.plot.scatter(x = 'weight', y = 'height');
```

```python
x = gym.weight
y = gym.height
plt.scatter(x,y)
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
x = iris.petal_length
y = iris.petal_width
plt.scatter(x,y)
plt.show()
```

```python
for name, group in iris.groupby("species"):
    plt.scatter(group['petal_length'], group['petal_width'], label=name)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.title('Scatter plot on Iris dataset')
plt.legend(loc=2)
plt.show()    
```

```python
fig, axs = plt.subplots(1, 3, sharex='col', sharey='row', gridspec_kw={'hspace': 0, 'wspace': 0.1}, figsize=(10,6))
fig.suptitle('Scatter plot on Iris dataset')
colors = ['r','g','b']
i = 0
for name, group in iris.groupby("species"):
    axs[i].scatter(group['petal_length'], group['petal_width'], color=colors[i])
    axs[i].set(xlabel=name)
    i += 1
plt.show()    
```
