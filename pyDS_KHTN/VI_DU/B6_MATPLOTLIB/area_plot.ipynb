```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Create data
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7] 
```

```python
# Area plot
plt.fill_between(x, y)
plt.show()
```

```python
# Create data
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]
 
# Change color, and its transparency
plt.fill_between( x, y, color="skyblue", alpha=0.5)
# Add a stronger line on top 
plt.plot(x, y, color="Slateblue", alpha=0.6)
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv('data/museum_visitors.csv', index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,7))
plt.fill_between(df.index, df['Avila Adobe'], label='Avila Adobe')
plt.fill_between(df.index, df['Firehouse Museum'], label='Firehouse Museum')
plt.fill_between(df.index, df['Chinese American Museum'], label='Chinese American Museum')
plt.fill_between(df.index, df['America Tropical Interpretive Center'], label='America Tropical Interpretive Center')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.fill_between(df.index, df['Avila Adobe'], label='Avila Adobe', alpha=0.5)
plt.fill_between(df.index, df['Firehouse Museum'], label='Firehouse Museum', alpha=0.5)
plt.fill_between(df.index, df['Chinese American Museum'], label='Chinese American Museum', alpha=0.5)
plt.fill_between(df.index, df['America Tropical Interpretive Center'], label='America Tropical Interpretive Center', alpha=0.5)
plt.plot(df.index, df['Avila Adobe'], alpha=0.6)
plt.plot(df.index, df['Firehouse Museum'], alpha=0.6)
plt.plot(df.index, df['Chinese American Museum'], alpha=0.6)
plt.plot(df.index, df['America Tropical Interpretive Center'], alpha=0.6)
plt.legend()
plt.grid()
plt.show()
```
