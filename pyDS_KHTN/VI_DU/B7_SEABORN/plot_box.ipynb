```python
# nói thêm catplot
tips = sns.load_dataset('tips') # category
```

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind='box');
```

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind='swarm');
```

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind='swarm', hue='smoker');
```

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind='swarm', hue='smoker', col='time');
```
