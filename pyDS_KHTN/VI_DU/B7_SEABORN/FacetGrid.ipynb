```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g2 = sns.FacetGrid(tips, row='time')
g2.map(sns.boxplot, 'total_bill')
plt.show()
```

```python
g2 = sns.FacetGrid(tips, col='time')
g2.map(sns.boxplot, 'total_bill')
plt.show()
```

```python
g2 = sns.FacetGrid(tips, col='sex')
g2.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
g2 = sns.FacetGrid(tips, col='day', col_wrap=2) # wrap column
g2.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.catplot(x="sex", y="age", hue="class", col="survived", data=titanic, 
            kind="strip"); # groups: x, hue, col, row
```

```python
sns.catplot(x="sex", y="age", hue="class", col="survived", data=titanic, 
            kind="bar", ci=None);
```

#### Categorical scatterplots:

- :func:`stripplot` (with ``kind="strip"``; the default)
- :func:`swarmplot` (with ``kind="swarm"``)

#### Categorical distribution plots:

- :func:`boxplot` (with ``kind="box"``)
- :func:`violinplot` (with ``kind="violin"``)
- :func:`boxenplot` (with ``kind="boxen"``)

#### Categorical estimate plots:

- :func:`pointplot` (with ``kind="point"``)
- :func:`barplot` (with ``kind="bar"``)
- :func:`countplot` (with ``kind="count"``)

```python
sns.factorplot(data=tips, kind='box', x='total_bill', col='sex');
# kind: point, bar, strip, swarm, box, violin, or boxen
```

```python
# point plot(by default) 
sns.factorplot(x ='size', y ='tip', data = tips)   
# Show the plot 
plt.show() 
```

```python
# point plot using hue attribute 
# for colouring out points 
# according to the sex 
sns.factorplot(x ='size', y ='tip', hue = 'sex', data = tips)   
# Show the plot 
plt.show() 
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.lmplot(x="petal_length", y="petal_width", data=iris, hue='species'); # linear model
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue='sex');
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue='sex', col='time');
```

```python
sns.PairGrid(iris, vars=['petal_length', 'petal_width']).map(sns.scatterplot);
```

```python
g = sns.PairGrid(iris, vars=['petal_length', 'petal_width'])
g.map_diag(sns.distplot)
g.map_offdiag(sns.scatterplot)
plt.show()
```

```python
sns.pairplot(iris, vars=['petal_length', 'petal_width'], kind='scatter',
            hue='species', palette='tab10', diag_kws={'alpha':0.7})
plt.show()
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g.plot(sns.regplot, sns.distplot)
plt.show()
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g.plot_joint(sns.scatterplot, s=100, alpha=.5)
g.plot_marginals(sns.distplot, kde=False)
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
plt.show()
```
