```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matplotlib.__version__ # matplotlib-3.3.2
```

```python
data = pd.read_csv("data/salary.csv", index_col=0)
```

```python
data.shape
```

```python
data.head()
```

```python
sns.scatterplot(x="Salary", y="Age", data=data);
```

```python
sns.scatterplot(x="Salary", y="Age", hue="Education", data=data);
```

```python
sns.relplot(x="Salary", y="Age", hue='Education', col="Gender", data=data);
```

```python
# vẽ tương tự cho tips
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue='sex', data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue='sex', col='time', data=tips);
```

```python
sns.set_style("whitegrid")
sns.relplot(x="total_bill", y="tip", hue='sex', col='time', data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue='sex', col='time', style='sex', data=tips);
```
