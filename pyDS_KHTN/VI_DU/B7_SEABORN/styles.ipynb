```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# pandas plot
iris['petal_length'].plot.hist()
plt.show()
```

```python
# set the default seaborn style
sns.set()
```

```python
iris['petal_length'].plot.hist()
plt.show()
```

```python
for style in ['white','dark','whitegrid','darkgrid','ticks']:
    sns.set_style(style)
    sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
    plt.show()
```

```python
sns.set(color_codes=True)
sns.distplot(iris['petal_length'], color='r');
```

```python
sns.set_palette('Dark2')
ax = sns.swarmplot(x='species', y='petal_length', data=iris)
ax.set_title('Iris types', fontsize=14, color='blue')
plt.show()
```

```python
p = sns.color_palette('pastel')
ax = sns.boxplot(x='species', y='petal_length', data=iris, palette=p)
ax.set_title('Iris types', fontsize=14, color='red')
plt.show()
```
