```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # set default theme
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips);
```

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue='sex', data=tips); # lm: linear model
```

```python
sns.lmplot(x="total_bill", y="tip", col='sex', data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue='sex', col='time', row='smoker', data=tips);
```
