```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset la tips
# ve stripplot voi x='time' va y='total_bill'
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.stripplot(x='time', y='total_bill', data=tips)
plt.show()
# số thực khách ăn vào buổi chiều nhiều hơn buổi trưa
# trong các buổi ăn thì buổi chiều có các số tiền lớn hơn
```

```python
sns.stripplot(x='time', y='total_bill', data=tips, hue='sex')
plt.show()
```

```python
sns.swarmplot(x='time', y='total_bill', data=tips, hue='smoker')
plt.show()
```

```python
sns.boxplot(x='time', y='total_bill', data=tips)
plt.grid()
plt.show()
```

```python
sns.boxplot(x='time', y='total_bill', data=tips)
sns.swarmplot(x='time', y='total_bill', data=tips, color='white')
plt.grid()
plt.show()
```
