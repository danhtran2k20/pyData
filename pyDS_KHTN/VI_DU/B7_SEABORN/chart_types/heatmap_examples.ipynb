```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/daily_show_guests_cleaned.csv') 
# we will start by looking at guests of the Daily Show from 1999 - 2015 
# and see how the occupations of the guests have changed over time.
```

```python
df.shape
```

```python
df.head()
```

```python
df_ = pd.crosstab(df['Group'], df['YEAR'])
```

```python
df_
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_, cbar=False, cmap='YlGnBu', linewidths=0.3)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

```python
# vẽ heatmap cho các chuyến bay deplay trong năm
# Path of the file to read
flight_filepath = "../data/flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
flight_data
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,8))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True, linewidths=0.5)

# Add label for horizontal axis
plt.xlabel("Airline");
```
