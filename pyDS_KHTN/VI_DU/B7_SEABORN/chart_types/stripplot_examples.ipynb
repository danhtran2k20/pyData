```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.stripplot(x="species", y="petal_length", data=iris, jitter=False);
```

```python
sns.stripplot(x="species", y="petal_length", data=iris, jitter=True);
```

```python
sns.stripplot(x="species", y="petal_length", data=iris, size=3);
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.stripplot(x="sex", y="tip", data=tips);
```

```python
sns.stripplot(y="sex", x="tip", data=tips);
```

```python
sns.stripplot(x="sex", y="tip", hue='smoker', data=tips);
```

```python
sns.stripplot(x="sex", y="tip", hue='size', data=tips);
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.stripplot(x="survived", y="age", data=titanic);
```

```python
sns.stripplot(x="pclass", y="age", data=titanic);
```
