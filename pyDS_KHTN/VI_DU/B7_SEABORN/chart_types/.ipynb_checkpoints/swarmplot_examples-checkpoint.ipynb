```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.swarmplot(x="species", y="petal_length", data=iris);
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.swarmplot(x="sex", y="tip", data=tips);
```

```python
sns.swarmplot(y="sex", x="tip", data=tips);
```

```python
sns.swarmplot(x="sex", y="tip", hue='smoker', data=tips);
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.swarmplot(x="survived", y="age", data=titanic);
```

```python
sns.swarmplot(x="alive", y="age", data=titanic);
```

```python
sns.swarmplot(x="pclass", y="age", data=titanic);
```

```python
sns.swarmplot(x="pclass", y="age", hue='survived', data=titanic);
```

```python
titanic.groupby(['pclass','survived']).size()
```

```python
sns.swarmplot(x="pclass", y="sex", hue='survived', data=titanic);
```
