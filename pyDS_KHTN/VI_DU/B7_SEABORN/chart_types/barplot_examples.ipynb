```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/salary.csv", index_col=0)
```

```python
data.head()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=data, x='Salary', y='Education')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=data, x='Salary', y='Education', hue='Gender')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=data, x='Education', y='Salary')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.pointplot(data=data, x='Education', y='Salary', capsize=0.1)
plt.show()
```
