```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("../data/salary.csv", index_col=0)
```

```python
data.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=data, x='Salary', y='Education')
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=tips, x='day', y='total_bill', palette='husl')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips, x='day', y='total_bill', palette='husl')
plt.show()
```

```python
# Violin plots are a combination of box plots and density plots. 
# Both the statistical measures and the distribution are visualized. 
# The thick black bar in the center represents the interquartile range, 
# the thin black line shows the 95% confidence interval, and the white dot shows the median.
# On both sides of the center-line, the density is visualized.
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips, x='day', y='total_bill', palette='husl')
sns.swarmplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips, x='day', y='total_bill', palette='pastel')
sns.boxplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips, x='day', y='total_bill', palette='husl', hue='sex')
plt.show()
```
