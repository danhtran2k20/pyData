```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g2 = sns.FacetGrid(tips, row='time')
g2.map(sns.boxplot, 'total_bill')
plt.show()
```

```python
g2 = sns.FacetGrid(tips, col='time')
g2.map(sns.boxplot, 'total_bill')
plt.show()
```

```python
g2 = sns.FacetGrid(tips, col='time', row='sex')
g2.map(sns.boxplot, 'total_bill')
plt.show()
```

```python
# ve scatter cho 2 bien x=total_bill va y=tip
# col=sex
g2 = sns.FacetGrid(tips, col='sex')
g2.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
# cho hue=smoker
g2 = sns.FacetGrid(tips, col='sex', hue='smoker')
g2.map(sns.scatterplot, 'total_bill', 'tip')
plt.legend()
plt.show()
```

```python
sns.catplot(x="time", y="total_bill", hue="smoker", col="sex", data=tips, kind="strip"); 
```

#### Categorical scatterplots:

- :func:`stripplot` (with ``kind="strip"``; the default)
- :func:`swarmplot` (with ``kind="swarm"``)

#### Categorical distribution plots:

- :func:`boxplot` (with ``kind="box"``)
- :func:`violinplot` (with ``kind="violin"``)
- :func:`boxenplot` (with ``kind="boxen"``)

#### Categorical estimate plots:

- :func:`pointplot` (with ``kind="point"``)
- :func:`barplot` (with ``kind="bar"``)
- :func:`countplot` (with ``kind="count"``)

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.catplot(data=titanic, kind='bar', x='sex', y='age', hue='class', col='survived', ci=None);
```

```python
sns.factorplot(data=tips, kind='box', x='total_bill', col='sex');
# kind: point, bar, strip, swarm, box, violin, or boxen
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.lmplot(x="petal_length", y="petal_width", data=iris, hue='species'); # linear model
```

```python
sns.lmplot(x="petal_length", y="petal_width", data=iris, col='species');
```

```python
sns.lmplot(x="petal_length", y="petal_width", data=iris, hue='species', fit_reg=False);
```

```python
sns.lmplot(x="petal_length", y="petal_width", data=iris, hue='species', scatter=False);
```

```python
g2 = sns.FacetGrid(tips, col='day', col_wrap=2)
g2.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
g = sns.PairGrid(data=tips, vars=['total_bill','tip'])
g.map(sns.scatterplot)
plt.show()
```

```python
g = sns.PairGrid(data=tips, vars=['total_bill','tip'])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
plt.show()
```

```python
# tim moi tuong quan cua hai bien age va fare (titanic)
sns.pairplot(data=titanic, vars=['age','fare'], kind='scatter', palette='RdBu', hue='sex',
             diag_kws={'alpha':0.7})
plt.show()
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.distplot)
plt.show()
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.regplot, sns.distplot)
plt.show()
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.regplot, sns.distplot)
plt.show()
```
