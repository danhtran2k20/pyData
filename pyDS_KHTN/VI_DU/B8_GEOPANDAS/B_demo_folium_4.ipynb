```python
provinces = gpd.read_file('data/province data.geojson')
provinces.head()
```

```python
cac_ten_tinh = ['Hậu Giang', 'Tiền Giang']
styles = [{'fillColor': '#228B22', 'color': '#228B22'}, {'fillColor': '#00FFFFFF', 'color': '#00FFFFFF'}]
folium.GeoJson(provinces.loc[provinces['ten_tinh']==cac_ten_tinh[0],'geometry'],style_function=lambda x:styles[0]).add_to(vn_map)
folium.GeoJson(provinces.loc[provinces['ten_tinh']==cac_ten_tinh[1],'geometry'],style_function=lambda x:styles[1]).add_to(vn_map)
vn_map
```

```python
df=pd.read_csv('data/data.csv')
df
```

```python
df_deleted_rows = df[df['Loại dự án']=='Đang phát triển']
df_deleted_rows
```

```python
df.drop(df_deleted_rows.index)
```

```python
geolocator = Nominatim(user_agent="Anonymous")
location = geolocator.geocode("Quy Nhơn, Việt Nam")

vn_map = folium.Map(location=[location.latitude, location.longitude], zoom_start=6,tiles='cartodbpositron')
markers = folium.map.FeatureGroup()

for lat, lng, in zip(df.latitude, df.longitude):
    markers.add_child(
        folium.features.Marker(
            [lat, lng],
        )
    )

vn_map.add_child(markers)
vn_map
```

```python
for lat,lan,name,elev in zip(df['latitude'],df['longitude'],df['Tỉnh'],df['Tổng MW']): 
    text = folium.Html('<b>'+name+'</b><br> Tổng MW:<b>'+str(elev)+'</b>', script=True)

    popup = folium.Popup(text, max_width=260)
    
    folium.Marker(location=[lat,lan],popup = popup, 
                  icon= folium.Icon(icon_color='yellow',icon = 'cloud')).add_to(vn_map) 
vn_map    
```
