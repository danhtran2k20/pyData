```python
import folium
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hcm_latitude = 10.762622 
hcm_longitude = 106.660172
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
hcm_map
```

```python
df_hcm = pd.read_excel('data/HCMC_location.xlsx', sheet_name="Location")
df_hcm.head()
```

### Tạo các Circle Marker

```python
# instantiate a feature group for the district in the dataframe
districts = folium.map.FeatureGroup()
# loop through the dictricts and add each to the dictrict feature group
for lat, lng in zip(df_hcm.Latitude, df_hcm.Longitude):
    districts.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        )
    )
# add dictricts to map
hcm_map.add_child(districts)
```

### Tạo các Popup

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
# add pop-up text to each marker on the map
for lat, lng, name in zip(df_hcm.Latitude, df_hcm.Longitude, df_hcm.Name):
    folium.Marker([lat, lng], popup=name).add_to(hcm_map)    

hcm_map
```

### Tạo các Circle Marker có size thay đổi

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
# tạo các Marker
# instantiate a feature group for the district in the dataframe
districts = folium.map.FeatureGroup()
# loop through the dictricts and add each to the dictrict feature group
for lat, lng, pop in zip(df_hcm.Latitude, df_hcm.Longitude, df_hcm.Population):
    districts.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=pop/50, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.3
        )
    )
# add dictricts to map
hcm_map.add_child(districts)
```

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
# tạo các Popup
# add pop-up text to each marker on the map
latitudes = list(df_hcm.Latitude)
longitudes = list(df_hcm.Longitude)
labels = list(df_hcm.Name)
pops = list(df_hcm.Population)

for lat, lng, label, pop in zip(latitudes, longitudes, labels, pops):
    tt = 'Số dân: '+str(pop)
    folium.Marker([lat, lng], popup=label, tooltip=tt).add_to(hcm_map)    
hcm_map
```

```python
def set_color(pop):
    if pop<=200:
        return 'blue'
    elif pop<=500:
        return 'green'
    else:
        return 'red'
```

```python
df_hcm['color'] = df_hcm.Population.map(lambda pop: set_color(pop))
```

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
# tạo các Popup
# add pop-up text to each marker on the map
latitudes = list(df_hcm.Latitude)
longitudes = list(df_hcm.Longitude)
labels = list(df_hcm.Name)
pops = list(df_hcm.Population)
colors = list(df_hcm.color)

for lat, lng, label, pop, color in zip(latitudes, longitudes, labels, pops, colors):
    tt = 'Dân số: '+str(pop)
    folium.Marker([lat, lng], popup=label, tooltip=tt, icon=folium.Icon(color=color)).add_to(hcm_map)    
hcm_map
```
